## 中心極限定理
simulation　平均値の分布

```python
import random
import matplotlib.pyplot as plt

N = [5, 10, 100]
for n in N:
    means = []
    for i in range(10000):
        mean = 0
        for j in range(n):
            mean += random.randint(1,6)
        mean /= n
        means.append(mean)
    plt.hist(means, bins=10)
    plt.title('N={}'.format(n))
    plt.xlim(1,6)
    #plt.ylim(0,1200)
    plt.xlabel('mean')
    plt.ylabel('freq')
    plt.show()
```

### 確率密度を正規分布(平均μ、分散σ2/N)で近似

```python
from scipy.stats import norm
import numpy as np

N = [5, 10, 100]
mu = 7/2
var = 35/12
X = np.arange(1,6,0.01)
for n in N:
    means = []
    for i in range(10000):
        mean = 0
        for j in range(n):
            mean += random.randint(1,6)
        mean /= n
        means.append(mean)
    plt.hist(means, bins=10, density=True)
    plt.title('N={}'.format(n))
    plt.xlim(1,6)
    #plt.ylim(0,1)
    plt.xlabel('mean')
    plt.ylabel('freq')
    
    
    Y = norm.pdf( X,mu, np.sqrt(var/n) )
    plt.plot(X,Y,color='r')
    plt.show()
```

```python
import numpy as np

f = open('../data/test_score_data.txt', 'r')

score = []
f.readline()
for line in f:
    item = line[:-1].split('\t')
    item = [int(num) for num in item[2:]]
    score.append(sum(item))
plt.hist(score, bins=10, density=True)

score = np.array(score)
mean = score.mean()
var = score.var()
X = np.arange(1,80,0.1)
Y = norm.pdf( X, mean, np.sqrt(var))
plt.plot(X,Y,color='r')
plt.xlabel('score')
plt.ylabel('Probability density')
plt.show()
```
